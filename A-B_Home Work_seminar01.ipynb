# ozon.ru

Ключевые метрики:

1. DAU - daily active users — пользовательская активность на веб-сайте за сутки.

2. WAU - weekly active users - пользовательская активность на веб-сайте за неделю.

3. MAU - monthly active users - пользовательская активность на веб-сайте за месяц.

4. CR - conversion rate - коэффициент конверсии — процент посетителей сайта, совершивших покупку.

5. ARPU - average revenue per user - средний доход с привлеченного пользователя.

6. CAC - cost per customer - стоимость приобретения нового клиента.

7. retention rate - ровень удержания клиентов — процент клиентов, которые возвращаются, чтобы снова совершить покупку.

8. NPS - net promoter score — показатель лояльности и удовлетворенности клиентов.

9. PPV - pages per visit — глубина просмотра страниц — cреднее время, которое посетитель проводит на сайте.

10. Churn rate - отток клиентов - процент посетителей, покидающих сайт.

11. Коэффициент отказа от корзины — процент посетителей, которые добавляют товары в корзину, но не завершают покупку.



Метрики дают представление о клиенте и общей производительности платформы. 

Выбранные метрики можно измерить с помощью различных фреймворков.
